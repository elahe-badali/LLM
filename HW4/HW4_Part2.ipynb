```
%%time

import torch
import numpy as np
import plotly.graph_objects as go
from tqdm import tqdm
import imageio
import os
from transformers import GPT2LMHeadModel, GPT2Tokenizer
# Load pre-trained model
model_name = 'gpt2'
model =  GPT2LMHeadModel.from_pretrained(model_name)# TODO
tokenizer = GPT2Tokenizer.from_pretrained(model_name)# TODO

# Set model to evaluation mode
# TODO
model.eval()

# Define our input
input_text = "I have a dream"
inputs =tokenizer(input_text, return_tensors="pt") # TODO encode the inputs

# Compute the original loss
outputs = model(**inputs, labels=inputs["input_ids"])
original_loss = outputs.loss.item() # TODO

# Define two orthogonal random directions for each model parameter
direction1 = [torch.randn_like(p) for p in model.parameters()]# TODO a list of directions for each model parameter
direction2 = [torch.randn_like(p) for p in model.parameters()] # TODO

# Normalize vectors
for p, d1, d2 in zip(model.parameters(), direction1, direction2):
    norm_p = torch.linalg.norm(p.flatten())
    d1.div_(torch.linalg.norm(d1.flatten())).mul_(norm_p)
    d2.div_(torch.linalg.norm(d2.flatten())).mul_(norm_p)

# Define the range to explore
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)

# Prepare to collect the losses
Z =  np.zeros_like(X)# TODO

# Compute loss for each direction
for i in tqdm(range(x.size), desc="x progress"):
    for j in tqdm(range(y.size), desc="y progress", leave=False):
        # Perturb the model parameters
        for p, d1, d2 in zip(model.parameters(), direction1, direction2):
            # TODO
            p.data.add_(x[i] * d1 + y[j] * d2)


        # Compute the loss
        outputs = model(**inputs, labels=inputs["input_ids"])
        loss = outputs.loss.item()
        Z[i, j] = loss # TODO

        # Revert the model parameters
        # TODO
        with torch.no_grad():
            for p, d1, d2 in zip(model.parameters(), direction1, direction2):
                p.data.add_(-x[i] * d1 - y[j] * d2)
```

```
# Check whether direction1 and direction2 are really orthogonal, using cosine similarity
# TODO
direction1_concatenated = torch.cat([d.flatten() for d in direction1])
direction2_concatenated = torch.cat([d.flatten() for d in direction2])

cosine_similarity = torch.dot(
    direction1_concatenated, direction2_concatenated) / (
        torch.norm(direction1_concatenated) * torch.norm(direction2_concatenated))
print(f"Cosine Similarity between direction1 and direction2: {cosine_similarity.item()}")
```

```
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.update_layout(scene=dict(zaxis=dict(range=[Z.min(), Z.max()])))
fig.show()
```
